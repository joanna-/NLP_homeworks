```python
def contain_digit(x):
    return re.search(r'\d', x) is not None

def remove_html(x):
    return re.sub("<[^>]*>", "", x)

def remove_linebreaks(x):
    return re.sub("-\n", "", x)
```

```python
def process_post_response(text, collection):
    for line in text.splitlines():
        if ':' in line:
            try:
                splitted = line.split(':', 1)[0].split()
                collection.append((splitted[0].lower(), splitted[1]))
            except:
                pass
    print("Collection size is: %d" % len(collection))
```

```python
import numpy as np
import math

def llr(k):
    return 2 * sum(k) * (H(k) - H([k[0] + k[1], k[2] + k[3]]) - H([k[0] + k[2], k[1] + k[3]]))

def table(bigram):
    
    return [
        bigrams[bigram],
        frequency_list[bigram[1]] - bigrams[bigram],
        frequency_list[bigram[0]] - bigrams[bigram],
        bigrams_num + bigrams[bigram] - frequency_list[bigram[1]] - frequency_list[bigram[0]]
    ]

def H(k):
    N = sum(k)
    return sum([el / N * math.log(el / N + int(el == 0)) for el in k])

logs = {}
for bigram in bigrams.keys():
    logs[bigram] = llr(table(bigram))
```

```python
def is_part_of_key(word, values):
    return word.split(':')[1] in values

def is_valid_key(k):
    nouns = ['subst', 'depr']
    adjs = ['adj', 'adja', 'adjp', 'adjc']
    return is_part_of_key(k[0], nouns) and (is_part_of_key(k[1], nouns) or is_part_of_key(k[1], adjs))

filtered = {k: v for k, v in logs.items() if is_valid_key(k)}
sorted(filtered.items(), key=lambda x: x[1], reverse=True)[:30]  
```
